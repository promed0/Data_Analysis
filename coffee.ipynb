```python
ax = sns.countplot(x = 'money', data = df, hue = 'cash_type')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
In the above graph most of the customer pay 37.72 $ by their card and lowest is 18.12 $
```

```python
pay_card = df.groupby(['card'],as_index= False)['money'].sum().sort_values(by = 'money', ascending=False).head(12)
# sns.barplot(x = 'Gender', y = 'Amount', data=sels_gen)
sns.set(rc={'figure.figsize':(35,15)})
sns.barplot(x= 'card', y = 'money', data = pay_card)
```

```python
here the card ANON-xxxx-xxxx-xx12 card pay most of the money on the coffee shop which amount are 2500$+
```

```python
ax = sns.countplot(x = 'money', data = df, hue = 'coffee_name')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
When the "Americano with milk" coffee sold while the coffee price 32.82 and amount of the highest coffee serve is 147.
```

```python
coffee_name = df.groupby(['coffee_name'],as_index= False)['money'].sum().sort_values(by = 'money', ascending=False)
# sns.barplot(x = 'Gender', y = 'Amount', data=sels_gen)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'coffee_name', y = 'money', data = coffee_name,hue=('coffee_name'))
```

```python
From the above Barchat we get the highest money earn from the Americano with milk and lowest income from the Espresso
```
