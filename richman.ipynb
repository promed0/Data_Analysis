```python
ax = sns.countplot( x = 'Country',data= df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
Char gives us, The most company in the USA and the lowest company in the INDIA
```

```python
ax = sns.countplot( x = 'Country',data= df,hue='Industry')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
USA have the most Technology industry and the number of the industry are 254 but lowest Technology company in the Mexico. Almost most of the industry 
in the USA. 
```

```python
ax = sns.countplot( x ='Name',data= df)
sns.set(rc={'figure.figsize':(40,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
From the chart we get the most Company share name is Larry Ellison total company are 67
```

```python
company_worth = df.groupby(['Country'],as_index=False)['Net Worth (in billions)'].sum().sort_values(by='Net Worth (in billions)', ascending=True).head(5)
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='Country',y='Net Worth (in billions)',data=company_worth,hue='Net Worth (in billions)')
```

```python
Most worth company in the USA then stand France and three other country are average Net worth.
```
