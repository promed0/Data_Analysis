```python
title_rating = df.groupby(['ID'], as_index=False)['Year'].sum().sort_values(by='Year', ascending=False).head(50)
# Set figure size
sns.set(rc={'figure.figsize': (20, 3)})

# Create a bar plot for top-rated movies by title and rating
sns.barplot(x='ID', y='Year', data=title_rating, hue='Year')
# plt.show()
```

```python
20 years experience employee in the company and lowest are the experience employe are below 18 years
```

```python
title_rating = df.groupby(['ID'], as_index=False)['Salary'].sum().sort_values(by='Salary', ascending=False).head(50)
# Set figure size
sns.set(rc={'figure.figsize': (20, 3)})

# Create a bar plot for top-rated movies by title and rating
sns.barplot(x='ID', y='Salary', data=title_rating, hue='Salary')
```

```python
from chart, Highest salary pick 250000 salary plus per year, are lowest salary are 20000 below.
```

```python
title_rating = df.groupby(['Gender'], as_index=False)['ID'].sum().sort_values(by='ID', ascending=False)
# Set figure size
sns.set(rc={'figure.figsize': (10, 5)})

# Create a bar plot for top-rated movies by title and rating
sns.barplot(x='Gender', y='ID', data=title_rating, hue='Gender')
```
